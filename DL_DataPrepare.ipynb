```python
import os
from google.colab import files
import requests
from bs4 import BeautifulSoup
import os
import time

# 카테고리 접미사(여기에서 변경)
category_suffix = 'ATBB'

# CPDL 카테고리 페이지 URL
category_url = f'https://www.cpdl.org/wiki/index.php/Category:{category_suffix}'

# 작곡가 페이지 링크를 저장할 리스트
composer_pages = []

# HTTP GET 요청을 보내 카테고리 페이지의 내용을 받아오기
response = requests.get(category_url)
soup = BeautifulSoup(response.text, 'html.parser')

# 모든 <a> 태그를 찾고, 작곡가 페이지로 가는 링크를 추출
for link in soup.find_all('a', href=True, title=True):
    if '/wiki/index.php/' in link['href']:
        composer_pages.append(f"https://www.cpdl.org{link['href']}")

# 각 작곡가 페이지로 들어가서 MIDI 파일을 찾기
for composer_page in composer_pages:
    try:
        # 작곡가 페이지로 요청
        response = requests.get(composer_page)
        soup = BeautifulSoup(response.text, 'html.parser')

        # MIDI 파일 링크를 찾기
        for midi_link in soup.find_all('a', href=True, class_="internal"):
            href = midi_link['href']
            if href.endswith('.MID'):
                midi_url = f"https://www.cpdl.org{href}"

                # MIDI 파일 다운로드
                midi_response = requests.get(midi_url)
                filename = href.split('/')[-1]

                # MIDI 파일을 저장 경로 설정
                save_directory = f'/content/drive/My Drive/DL_Dataset/{category_suffix}'
                if not os.path.exists(save_directory):
                    os.makedirs(save_directory)

                # MIDI 파일을 저장
                with open(os.path.join(save_directory, filename), 'wb') as file:
                    file.write(midi_response.content)
                    print(f'Downloaded {filename}')

                # 지연 시간 설정
                time.sleep(2)
    except Exception as e:
        print(f'An error occurred while processing {composer_page}: {e}')
```
